 tanh = $\frac{e^x - e^ -x}{e^x + e^ -x}$

```python
# Lets import the dependencies first
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lets initialize a simple tensor to apply these functions and see
A = tf.cast(tf.range(-25, 25), tf.float32)
''' Casting because I want the range tensor to be of float datatype, so after appplying the functions,
I get point values '''
A
```

# 1. Sigmoid

```python
def sigmoid(A):
  return 1 / (1+ np.exp(-A))
```

```python
sigmoid(A)
```

```python
# Lets visualize
plt.plot(sigmoid(A))
```

# 2. Tanh

```python
def tanh(A):
  return (np.exp(A) - np.exp(-A)) / (np.exp(A) + np.exp(-A))
```

```python
tanh(A)
```

```python
# Lets visualize
plt.plot(tanh(A))
```

# 3. ReLU

```python
def relu(A):
  return np.maximum(0, A)
```

```python
relu(A)
```

```python
# Visualize
plt.plot(relu(A))
```

# 4. Leaky ReLU

```python
def leaky_relu(A):
  return np.maximum(0.1*A, A)
```

```python
leaky_relu(A)
```

```python
plt.plot(leaky_relu(A))
```

# 5. ELU

we do not consider Maxout yet, because it would need weights and biases, but the picture is clear enough to help implementing with weights and biases
* $\alpha$ is the learning rate.

```python
def ELU(A, alpha):
  array = []
  for i in A:
    if ( i < 0):
       array.append(alpha * (np.exp(i) - 1))
    else:
      array.append(i)
  return array
```

```python
ELU (A, 0.01) # Considering learning rate to be 0.01 for example, which is also by default
```

```python
# Lets plot and see
plt.plot(ELU(A, 0.01))
```
